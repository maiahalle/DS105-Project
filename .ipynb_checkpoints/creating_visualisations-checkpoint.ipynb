```python
g_republicans = (
    ggplot(df_republicans.head(50), aes(x="Word",y="Count")) + geom_col() + coord_flip()
)
```

```python
g_democrats = (
    ggplot(df_democrats.head(50), aes(x="Word",y="Count")) + geom_col() + coord_flip()
)
```

```python
g_democrats
```

### Word clouds:

```python
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS
```

```python
dtext = " ".join(i for i in df_democrats.Word)
dwordcloud = WordCloud(background_color="white").generate(dtext)
plt.figure( figsize=(15,10))
plt.imshow(dwordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
plt.savefig('democrats_word_cloud.png')
```

```python
rtext = " ".join(i for i in df_republicans.Word)
rwordcloud = WordCloud(background_color="white").generate(rtext)
plt.figure( figsize=(15,10))
plt.imshow(rwordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
plt.savefig('republicans_word_cloud.png')
```

## Loading more data:

```python
df_republican_dates_tweets = pd.read_csv("republican_dates_tweets.csv")
```

```python
df_republican_dates_tweets
```

```python
text = " ".join(item for item in df_republican_dates_tweets['tweet_text'])
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["https", "RT", "t", "co", "amp", "will", "s", "U"])
```

```python
wordcloud = WordCloud(background_color="white", stopwords = stopwords).generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

```python
democrat_words = ''
for i in range(len(df_democrat_keywords)):
    for word in df_democrat_keywords.iloc[i]['key_word_list']:
        democrat_words += word
```
